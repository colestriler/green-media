# Polarity in Headlines
It turns out that we can use the words in news headlines to calculate a measure of the sentiment of the headline. For example, the sentence "I love America!" has positive sentiment, whereas the sentence "I hate taxes!" has a negative sentiment. In addition, some words have stronger positive / negative sentiment than others: "I love America." is more positive than "I like America."

We will use the VADER (Valence Aware Dictionary and sEntiment Reasoner) lexicon to analyze the sentiment of the news headlines. VADER is a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social & news media which is great for our usage.

Polarity is defined as:

<li> $Polarity > 0$ Means **Positive** News Headlines

<li> $Polarity < 0$ Means **Negative** News Headlines

<li> $Polarity = 0$ = Means **Neutral** News Headlines
