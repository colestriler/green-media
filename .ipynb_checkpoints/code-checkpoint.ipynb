## Finding Common Words in Headlines
We will begin by looking at the most common words in the article headlines. Looking at the most common words may give us some insight into **FILL IN**

```python
def eliminate_excess(series):
    """In news headlines, there are many excess punctuation.
    This function removes all of that."""
    punct_re = r'[^A-Za-z\d\s]'
    #series = series.apply(lambda x: x.replace("", ""))
    series = series.str.replace(punct_re, " ")
    return series.str.lower()
```

```python
def count_frame(df, n):
    """Takes in a data frame and int n:
    
    DataFrame/ df: dataframe with HEADLINE column
    int/ n: number of words to choose
    r/type: 2 column dataframe most common words & counts"""
    
    counts = to_dict(df.headline)
    frame = pd.DataFrame.from_dict(counts, orient='index')
    frame = frame.sort_values(0, ascending=False).reset_index().head(n)
    frame = frame.rename(columns={'index': 'word', 0:'count'})
    return frame
```

```python
# cleaning columns
news.loc[:, 'headline'] = eliminate_excess(news['headline'])
news.loc[:, 'short_description'] = eliminate_excess(news['short_description'])

# making new DataFrames
env = news[(news['category']=='ENVIRONMENT')]
green = news[news['category']=='GREEN']
env_and_green = news[(news['category']=='GREEN') | (news['category']=="ENVIRONMENT")]

f"""There are {env.shape[0]} ENVIRONMENT articles and {green.shape[0]} GREEN articles"""
```

```python
# words that we will exclude from our model
stopwords = ["ourselves", "hers", "between", "yourself", "but", "again", "there", "about", "once", 
             "during", "out", "very", "having", "with", "they", "own", "an", "be", "some", "for", 
             "do", "its", "yours", "such", "into", "of", "most", "itself", "other", "off", "is", "s", 
             "am", "or", "who", "as", "from", "him", "each","the", "themselves", "until", "below", 
             "are", "we", "these", "your", "his", "through", "don", "nor", "me", "were", "her", "more", 
             "himself", "this", "down", "should", "our", "their", "while", "above", "both", "up", "to", 
             "ours", "had", "she", "all", "no", "when", "at", "any", "before", "them", "same", "and", 
             "been", "have", "in", "will", "on", "does", "yourselves", "then", "that", "because", "what", 
             "over", "why", "so", "can", "did", "not", "now", "under", "he", "you", "herself", "has", 
             "just", "where", "too", "only", "myself", "which", "those", "i", "after", "few", "whom", "t", 
             "being", "if", "theirs", "my", "against", "a", "by", "doing", "it", "how", "further", "was", 
             "here", "than", "--"]
```

```python
def to_dict(series):
    """Takes in a pd.Series of Strings and
        1) Takes every string
        2) Splits into individual words
        3) Checks if word is a stopword
        4) if not a stopword, add it to a dictionary
        5) If it is already in the dictionary, add a count"""
    dic = {}
    array = series.values
    for num in range(len(array)):
        headline = array[num].split()
        for word in headline:
            if word not in stopwords:
                if word in dic:
                    dic[word] += 1
                else:
                    dic[word] = 0
    return dic
    
```

# News Polarity

```python
index = pd.read_table('vader_lexicon.txt', header = None).iloc[:, 0]
polarity = {'polarity': pd.read_table('vader_lexicon.txt', header = None).iloc[:, 1]}
sent = pd.DataFrame(data = polarity).set_index(index)
sent.index.name = None

assert isinstance(sent, pd.DataFrame)
assert sent.shape == (7517, 1)
assert list(sent.index[5000:5005]) == ['paranoids', 'pardon', 'pardoned', 'pardoning', 'pardons']
assert np.allclose(sent['polarity'].head(), [-1.5, -0.4, -1.5, -0.4, -0.7])

punct_re = r'[^A-Za-z\d\s]'
news['hdl_no_punct'] = news['headline'].str.replace(punct_re, " ")
news['desc_no_punct'] = news['short_description'].str.replace(punct_re, " ")

tidy_format = news['desc_no_punct'].str.split(expand=True).stack().to_frame().reset_index(level = 1)
tidy_format.columns = ["num", "word"]
tidy_format.head()
```

```python
# Adding Polarity to NEWS table
news.loc[:, 'polarity'] = (tidy_format.merge(sent, left_on = 'word', 
                                       right_index = True)['polarity']
                                        .reset_index().groupby('index').sum())
news.loc[:, 'polarity'].replace(np.float64('nan'), 0, inplace = True)

# Adding Polarity to ENVIRONMENT table
env.loc[:, 'polarity'] = (tidy_format.merge(sent, left_on = 'word', 
                                       right_index = True)['polarity']
                                        .reset_index().groupby('index').sum())
env.loc[:, 'polarity'].replace(np.float64('nan'), 0, inplace = True)

# Adding Polarity to GREEN table
green.loc[:, 'polarity'] = (tidy_format.merge(sent, left_on = 'word', 
                                       right_index = True)['polarity']
                                        .reset_index().groupby('index').sum())
green.loc[:, 'polarity'].replace(np.float64('nan'), 0, inplace = True)

# Adding Polarity to ENV_AND_GREEN table
# Adding Polarity to GREEN table
env_and_green.loc[:, 'polarity'] = (tidy_format.merge(sent, left_on = 'word', 
                                       right_index = True)['polarity']
                                        .reset_index().groupby('index').sum())
env_and_green.loc[:, 'polarity'].replace(np.float64('nan'), 0, inplace = True)
```

```python
# taking the 100 most positive/negative headlines and putting in dictionary
most_positive = env_and_green.sort_values('polarity', ascending=False)
most_negative = env_and_green.sort_values('polarity', ascending=True)

most_positive_counts = count_frame(most_positive, 20)
most_negative_counts = count_frame(most_negative, 20)
```

## Let's See if We Can Guess the Political Bias of Hiffington Post

```python
# make new dataframes for TRUMP & OBAMA
trump = env_and_green[env_and_green['headline'].str.contains("trump")]
obama = env_and_green[env_and_green['headline'].str.contains("obama")]
presidents = env_and_green[(env_and_green['headline'].str.contains("obama")) | env_and_green['headline'].str.contains("trump")]
presidents.loc[:, "contains_trump"] = presidents['headline'].str.contains('trump')

# mean polarity of each president
trump_mean = '%.3f'%np.mean(trump.polarity)
obama_mean = '%.3f'%np.mean(obama.polarity)
```

```python
plt.figure(figsize=(15,7))
sns.distplot(trump['polarity'], hist=False, label=f"Trump (mean = {trump_mean})", rug=True)
sns.distplot(obama['polarity'], hist=False, label=f"Obama (mean = {obama_mean})", rug=True)
plt.title("Polarity of Articles that Contain The Words TRUMP vs OBAMA", fontsize=18)
plt.xlabel("Polarity", fontsize=18)
plt.xticks(fontsize=15)
plt.ylabel("Pecent (%)", fontsize=18)
plt.legend(fontsize=18, loc=2)
plt.savefig("img/political_bias.jpg")
plt.show()
```

As you can see, articles that contain the word "OBAMA" have an average polarity of 0.936 (more positive) and articles that contain the word "TRUMP" have an average polarity of -0.036 (slightly more negative). Therefore, we predict that the Huffington Post is a more liberal news source. It turns out, after a quick [Google search](https://www.google.com/search?ei=H6_MXMbkCNbn-gTotIS4Ag&q=huffington+post+political+leaning&oq=huffington+post+political&gs_l=psy-ab.1.1.0l4.2468.5312..6648...0.0..0.76.1408.25......0....1..gws-wiz.......0i71j0i131j0i131i67j0i67.8p1LHwdaKow), we were right. The Huffington Post is a leftward-leaning news source.
